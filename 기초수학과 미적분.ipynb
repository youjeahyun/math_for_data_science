```python
from sympy import *

i,n = symbols('i n')

# 1부터 n까지 i를 반복하면서
# 2를 곱하고 더합니다
summation = Sum(2*i,(i,1,n))

# n을 5로 지정하면
# 숫자 1에서 5까지 반복합니다
up_to_5 = summation.subs(n, 5)
print(up_to_5.doit())
```

```python
from math import exp

p = 100 #빌려준 금액
r = .20 #이자
t = 2.0 #기간
n = 12 #연간복리횟수

a = p * (1 + (r/n))**(n * t)

print(a)
```

```python
#연속 이자 계산하기 오일러의 수

from math import exp

p = 100 # 원금, 시작 금액
r = .20 # 연이율
t = 2.0 # 기간, 연도

a = p * exp(r*t)

print(a)
```
